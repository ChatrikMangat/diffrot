```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Loading the files
seq_k = np.loadtxt('dd2_s2_diffkep.d')
seq_s = np.loadtxt('dd2_s2_static.d')
seq_1 = np.loadtxt('dd2_s2_2.01m.d')
seq_2 = np.loadtxt('dd2_s2_2.30m.d')
seq_3 = np.loadtxt('dd2_s2_2.51m.d')
seq_4 = np.loadtxt('dd2_s2_2.74m.d')
seq_5 = np.loadtxt('dd2_s2_2.81m.d')
seq_6 = np.loadtxt('dd2_s2_2.85m.d')
seq_7 = np.loadtxt('dd2_s2_2.91m.d')

#Kepler
j_kep = seq_k[:,3]
f_kep = seq_k[:,5]
tw_kep = seq_k[:,8]
mg_kep = seq_k[:,4]
rho_kep = seq_k[:,12]
mb_kep = seq_k[:,9]

#Static
j_sta = seq_s[:,3]
f_sta = seq_s[:,5]
tw_sta = seq_s[:,8]
mg_sta = seq_s[:,4]
rho_sta = seq_s[:,12]
mb_sta = seq_s[:,9]

#Mb=1.287
j_1 = seq_1[:,3]
f_1 = seq_1[:,5]
tw_1 = seq_1[:,8]
mg_1 = seq_1[:,4]
rho_1 = seq_1[:,12]
mb_1 = seq_1[:,9]

#Mb=1.712
j_2 = seq_2[:,3]
f_2 = seq_2[:,5]
tw_2 = seq_2[:,8]
mg_2 = seq_2[:,4]
rho_2 = seq_2[:,12]
mb_2 = seq_2[:,9]

#Mb=2.177
j_3 = seq_3[:,3]
f_3 = seq_3[:,5]
tw_3 = seq_3[:,8]
mg_3 = seq_3[:,4]
rho_3 = seq_3[:,12]
mb_3 = seq_3[:,9]

#Mb=2.415
j_4 = seq_4[:,3]
f_4 = seq_4[:,5]
tw_4 = seq_4[:,8]
mg_4 = seq_4[:,4]
rho_4 = seq_4[:,12]
mb_4 = seq_4[:,9]

#Mb=2.692
j_5 = seq_5[:,3]
f_5 = seq_5[:,5]
tw_5 = seq_5[:,8]
mg_5 = seq_5[:,4]
rho_5 = seq_5[:,12]
mb_5 = seq_5[:,9]

#Mb=2.866
j_6 = seq_6[:,3]
f_6 = seq_6[:,5]
tw_6 = seq_6[:,8]
mg_6 = seq_6[:,4]
rho_6 = seq_6[:,12]
mb_6 = seq_6[:,9]

#Mb=2.966
j_7 = seq_7[:,3]
f_7 = seq_7[:,5]
tw_7 = seq_7[:,8]
mg_7 = seq_7[:,4]
rho_7 = seq_7[:,12]
mb_7 = seq_7[:,9]
```

```python
plt.figure(figsize=(12,8))
plt.plot(rho_kep,mg_kep,label='Kepler')
plt.plot(rho_sta,mg_sta,label='Static')
plt.plot(rho_1,mg_1,label='$M_b$ = 2.01')
plt.plot(rho_2,mg_2,label='$M_b$ = 2.30')
plt.plot(rho_3,mg_3,label='$M_b$ = 2.51')
plt.plot(rho_4,mg_4,label='$M_b$ = 2.74')
plt.plot(rho_5,mg_5,label='$M_b$ = 2.81')
plt.plot(rho_6,mg_6,label='$M_b$ = 2.85')
plt.plot(rho_7,mg_7,label='$M_b$ = 2.91')
plt.title('Differentially rotating stars at a=0.2')
plt.xlabel("log(rho)",size=15)
plt.ylabel("$M_g$ ($M_0$)",size=15)
plt.legend()
plt.grid('on')
plt.savefig('combined_mass.png')
```
